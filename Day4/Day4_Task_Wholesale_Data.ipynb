```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('wholesale_customers_data.csv')
data.head()
```

```python
data_new=pd.DataFrame(data.iloc[:,[False,False,True,True,True,True,True,True]])
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_new, method='ward'))
```

```python
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_new, method='ward'))
plt.axhline(y=200000, color='r', linestyle='--')
```

```python
#Agglomerative 3 Clusters
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')  
cluster.fit_predict(data_new)
```

```python
#KMeans 3 Clusters
from sklearn.cluster import KMeans
kmeanModel=KMeans(n_clusters=3)
kmeanModel.fit_predict(data_new)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(data_new['Milk'], data_new['Grocery'], c=cluster.labels_) 
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(data_new['Milk'], data_new['Grocery'], c=kmeanModel.labels_) 
```
